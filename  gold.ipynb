```python
import requests
from bs4 import BeautifulSoup
import pymysql
import pandas as pd
import csv
from urllib.request import urlopen
import threading
import schedule
import time
```

```python
def get():
    
    while True:
        html = urlopen('http://goldpricez.com/eg/gram')
        soup = BeautifulSoup(html, "html.parser")
        table = soup.findAll('table',attrs={'id':"gold_price_table2"})[0]
        rows = table.findAll("tr")

        with open("Price.csv", "wt+", newline="") as f:
            writer = csv.writer(f)
            for row in rows:
                csv_row = []
                for cell in row.findAll(["td", "th"]):
                    csv_row.append(cell.get_text())
                writer.writerow(csv_row)
        time.sleep(1440)            
get()          
```
